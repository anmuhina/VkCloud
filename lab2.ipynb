```python
#9. Выведите строки DataFrame, где уровень концентрации оксида углерода (CO(GT))превышает 3 мг/м³, а температура (T) ниже 20°C
df['CO(GT)'] = df['CO(GT)'].str.replace(',', '.')
df['CO(GT)'] = pd.to_numeric(df['CO(GT)'], errors='coerce')
df['T'] = df['T'].str.replace(',', '.')
df['T'] = pd.to_numeric(df['T'], errors='coerce')
df['C6H6(GT)'] = df['C6H6(GT)'].str.replace(',', '.')
df['RH'] = df['RH'].str.replace(',', '.')
df['AH'] = df['AH'].str.replace(',', '.')

print(df[(df['CO(GT)'] > 3) & (df['T'] < 20)])
```

```python
#10. Создайте новый признак "NOx_CO_ratio", который будет представлять собой отношение уровня концентрации оксидов азота (NOx(GT)) к уровню концентрации оксида углерода (CO(GT))
df['NOx_CO_ratio'] = df['NOx(GT)'] / df['CO(GT)']
df
```

```python
#13. Найдите количество строк, в которых отсутствуют данные о концентрации бензола (C6H6(GT)). Выведите информацию об этих строках. (пропущенные значения могут быть представлены как -200)
missing_list = df[df['C6H6(GT)'].isnull() | (df['C6H6(GT)'] == -200)]
print(missing_list, "\n")
print("Кол-во строк с пропусками:", missing_list.shape[0])
```

```python
#14. Найдите минимальный уровень концентрации оксида углерода (CO(GT)) при температуре выше 25°C. Выведите информацию об этих измерениях
min_CO_above_25 = df[df['T'] > 25]['CO(GT)'].min()
min_CO_info = df[(df['T'] > 25) & (df['CO(GT)'] == min_CO_above_25)]
print("Инфа:", min_CO_info)
print("Мин уровень CO при T > 25°C:", min_CO_above_25)
```

```python
#15. Найдите количество измерений, в которых относительная влажность (RH) превышает 90%. Выведите информацию об этих измерениях
df['RH'] = pd.to_numeric(df['RH'], errors='coerce')
rh_over_90 = df[df['RH'] > 90]
count = rh_over_90.shape[0]
print("Инфа:", rh_over_90)
print("Количество измерений с RH > 90%:", count)
```

```python
#16. Найдите разницу между средним уровнем концентрации оксида углерода (CO(GT)) при температуре выше и ниже 20°C. Округлите результат до двух знаков после запятой
mean_CO_above_20 = df[df['T'] > 20]['CO(GT)'].mean()
mean_CO_below_20 = df[df['T'] <= 20]['CO(GT)'].mean()
diff = round(mean_CO_above_20 - mean_CO_below_20, 2)
print("Уровень концентрации при температуре ниже 20:", mean_CO_below_20)
print("Уровень концентрации при температуре выше 20:", mean_CO_above_20)
print("\nРазница:", abs(diff))
```

```python
#17. Создайте признак "High_Ozone", который будет равен 1, если уровень концентрации озона (PT08.S5(O3)) превышает среднее значение, и 0 в противном случае
ozone_mean = df['PT08.S5(O3)'].mean()
df['High_Ozone'] = (df['PT08.S5(O3)'] > ozone_mean).astype(int)
df
```

```python
#19. Найдите количество измерений, в которых и уровень концентрации бензола (C6H6(GT)), и уровень концентрации оксидов азота (NOx(GT)) превышают свои средние значения
df['C6H6(GT)'] = pd.to_numeric(df['C6H6(GT)'], errors='coerce')
mean_benz = df['C6H6(GT)'].mean()
mean_nox = df['NOx(GT)'].mean()
count = df[(df['C6H6(GT)'] > mean_benz) & (df['NOx(GT)'] > mean_nox)].shape[0]
count
```

```python
#21. Найдите количество измерений, в которых уровень концентрации оксида углерода (CO(GT)) выше среднего значения по датасету. Выведите информацию об этих измерениях
mean_CO = df['CO(GT)'].mean()
above_mean_CO = df[df['CO(GT)'] > mean_CO]
count = above_mean_CO.shape[0]
print("Инфа:", above_mean_CO)
print("Количество измерений с CO(GT) выше среднего:", count)
```

```python
#22. Разделите данные на две группы: измерения, сделанные при температуре выше среднего значения, и измерения, сделанные при температуре ниже среднего значения. Сравните средний уровень концентрации бензола (C6H6(GT)) в этих двух группах
mean_temp = df['T'].mean()
above_mean_temp = df[df['T'] > mean_temp]
below_mean_temp = df[df['T'] <= mean_temp]
mean_benz_above = above_mean_temp['C6H6(GT)'].mean()
mean_benz_below = below_mean_temp['C6H6(GT)'].mean()
print("Средний уровень C6H6(GT) при температуре выше среднего:", mean_benz_above)
print("Средний уровень C6H6(GT) при температуре ниже среднего:", mean_benz_below)
```
